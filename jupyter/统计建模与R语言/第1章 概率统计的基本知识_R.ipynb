```r
myfun <- function(a) {
  x <- 1:100              #先生成一个1到100的序列，后面可以更改这些值，相当于覆盖掉原来的值
  x <- data.frame(x)
  a<-data.frame(a)
  for(i in 1:100){        #设置循环，循环抽取100个样本，并将计算出来的均值赋值给数据框中的x变量
    c<-a[sample(nrow(a),1000),]  #注意nrow()只用在data.frame
    m=mean(c)
    x$x[i]<-m  #覆盖掉x
  }
  par(mfrow=c(1,2))
  par(family = "STKaiti")
  plot(density(a$a), main = "原本分布") #destiny是核函数密度估计
  plot(density(x$x), main = "样本均值分布")  
}
```

```r
# 1.正态分布
options (warn = -1)
a<-rnorm(10000,0,1)
myfun(a)
```

```r
# 2.指数分布
b<-rexp(100000,5)
myfun(b)
```

```r
# 3.t分布
c<-rt(10000,3)
myfun(c)
```

```r
# 4.卡方分布
d<-rchisq(100000,5)
myfun(d)
```

```r
# 5.F分布
e <- rf(100000, 3, 20)
myfun(e)
```
