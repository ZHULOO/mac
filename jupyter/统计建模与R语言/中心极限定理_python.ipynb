###  中心极限定理模拟

#### 一. 生成数据(均匀分布)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 总体N = 10000
data = np.random.rand(100000)
sns.histplot(data,kde = True)
plt.show()
```

#### 二. 抽取样本(样本容量为n,各抽取1000次)

###  1. n = 2

```python
# 图1
sample_mean =[]
for i in range(1, 1000):
    s = np.random.choice(data, size=2).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=2") 
plt.show()
```

### 2. n = 5

```python
# 图2
sample_mean =[]
for i in range(1, 1000):
    s = np.random.choice(data, size=5).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=5")
plt.show()
```

### 3. n = 10

```python
# 图3
sample_mean =[]
for i in range(1, 1000):
    s = np.random.choice(data, size=10).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=10")
plt.show()
```

### 4. n = 30

```python
# 图4
sample_mean =[]
for i in range(1, 1000):
    s = np.random.choice(data, size=30).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=30")
plt.show()
```

#### 三. 总体服从其它分布

### 1. 三角形分布

```python
np.random.seed(123456)
data = np.random.triangular(1,5,10,10000) 
sns.histplot(data,kde = True)
plt.show()
```

### 2. 均匀分布

```python
data = numpy.random.uniform(1,100,10000)
sns.histplot(data,kde = True)
plt.show()
```

### 3. 正态分布

```python
data = np.random.normal(0,1,100000)
sns.histplot(data,kde = True)
plt.show()
```

### 4. 泊松分布

```python
data = numpy.random.poisson(2,10000)
sns.histplot(data,kde = True)
plt.show()
```

### 5. beta分布(0-1)

```python
data = numpy.random.beta(0.8,0.2,10000) #(p,q)
sns.histplot(data,kde = True) 
plt.show()
```

### 6. 二项分布

```python
data = numpy.random.binomial(10,0.3,10000) # (n,p)
sns.histplot(data,kde = True)
plt.show()
```

### 7. 卡方分布

```python
data = np.random.chisquare(30,10000)
sns.histplot(data,kde = True)
plt.show()
```

### 8. t分布

```python
data = np.random.standard_t(5,10000)
sns.histplot(data,kde = True)
plt.show()
```

### 9. F分布

```python
data = np.random.f(50,80,10000)
sns.histplot(data,kde = True)
plt.show()
```

### 总体服从某种分布，从这一总体中抽取的简单随机样本均值分布

```python
# 生成随机总体
data = np.random.chisquare(3,10000) # 修改分布类型
plt.figure(figsize = (15,9))
# 图1
plt.subplot(221)
sample_mean =[]
for i in range(1, 500):
    s = np.random.choice(data, size=2).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=2") 
# 图2
plt.subplot(222)
sample_mean =[]
for i in range(1, 500):
    s = np.random.choice(data, size=5).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=5")
# 图3
plt.subplot(223)
sample_mean =[]
for i in range(1, 500):
    s = np.random.choice(data, size=10).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=10")
# 图4
plt.subplot(224)
sample_mean =[]
for i in range(1, 500):
    s = np.random.choice(data, size=50).mean()
    sample_mean.append(s)
sns.histplot(sample_mean,kde = True)
plt.title("Size=30")
plt.show()
```
