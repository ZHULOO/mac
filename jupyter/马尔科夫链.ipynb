## 2. é©¬å°”ç§‘å¤«é“¾æ¨¡å‹çŠ¶æ€è½¬ç§»çŸ©é˜µçš„æ€§è´¨
+ å‡è®¾æˆ‘ä»¬å½“å‰è‚¡å¸‚çš„æ¦‚ç‡åˆ†å¸ƒä¸ºï¼š[0.3,0.4,0.3],å³30%æ¦‚ç‡çš„ç‰›å¸‚ï¼Œ40%æ¦‚ç‡çš„ç†Šç›˜ä¸30%çš„æ¨ªç›˜ã€‚ç„¶åè¿™ä¸ªçŠ¶æ€ä½œä¸ºåºåˆ—æ¦‚ç‡åˆ†å¸ƒçš„åˆå§‹çŠ¶æ€$t_0$ï¼Œå°†å…¶å¸¦å…¥è¿™ä¸ªçŠ¶æ€è½¬ç§»çŸ©é˜µè®¡ç®—$t_1,t_2,t_3,...$çš„çŠ¶æ€ã€‚ä»£ç å¦‚ä¸‹ï¼š

```python
import numpy as np
matrix = np.matrix([[0.9,0.075,0.025],[0.15,0.8,0.05],[0.25,0.25,0.5]], dtype=float)
vector1 = np.matrix([[0.3,0.4,0.3]], dtype=float)
for i in range(100):
    vector1 = vector1*matrix
    print("Current round:" , i+1)
    print(vector1)
```

+ æˆ‘ä»¬ç°åœ¨æ¢ä¸€ä¸ªåˆå§‹æ¦‚ç‡åˆ†å¸ƒè¯•ä¸€è¯•ï¼Œç°åœ¨æˆ‘ä»¬ç”¨[0.7,0.1,0.2]ä½œä¸ºåˆå§‹æ¦‚ç‡åˆ†å¸ƒï¼Œç„¶åè¿™ä¸ªçŠ¶æ€ä½œä¸ºåºåˆ—æ¦‚ç‡åˆ†å¸ƒçš„åˆå§‹çŠ¶æ€$t_0$ï¼Œå°†å…¶å¸¦å…¥è¿™ä¸ªçŠ¶æ€è½¬ç§»çŸ©é˜µè®¡ç®—$t_1,t_2,t_3,...$çš„çŠ¶æ€ã€‚ä»£ç å¦‚ä¸‹ï¼š

```python
matrix = np.matrix([[0.9,0.075,0.025],[0.15,0.8,0.05],[0.25,0.25,0.5]], dtype=float)
vector1 = np.matrix([[0.7,0.1,0.2]], dtype=float)
for i in range(100):
    vector1 = vector1*matrix
    print("Current round:" , i+1)
    print(vector1)
```

+ è¿™ä¸ªæ€§è´¨ä¸å…‰å¯¹æˆ‘ä»¬ä¸Šé¢çš„çŠ¶æ€è½¬ç§»çŸ©é˜µæœ‰æ•ˆï¼Œå¯¹äºç»å¤§å¤šæ•°çš„å…¶ä»–çš„é©¬å°”ç§‘å¤«é“¾æ¨¡å‹çš„çŠ¶æ€è½¬ç§»çŸ©é˜µä¹Ÿæœ‰æ•ˆã€‚åŒæ—¶ä¸å…‰æ˜¯ç¦»æ•£çŠ¶æ€ï¼Œè¿ç»­çŠ¶æ€æ—¶ä¹Ÿæˆç«‹ã€‚
+ åŒæ—¶ï¼Œå¯¹äºä¸€ä¸ªç¡®å®šçš„çŠ¶æ€è½¬ç§»çŸ©é˜µğ‘ƒï¼Œå®ƒçš„næ¬¡å¹‚$p^n$åœ¨å½“nå¤§äºä¸€å®šçš„å€¼çš„æ—¶å€™ä¹Ÿå¯ä»¥å‘ç°æ˜¯ç¡®å®šçš„ï¼Œæˆ‘ä»¬è¿˜æ˜¯ä»¥ä¸Šé¢çš„ä¾‹å­ä¸ºä¾‹ï¼Œè®¡ç®—ä»£ç å¦‚ä¸‹ï¼š

```python
matrix = np.matrix([[0.9,0.075,0.025],[0.15,0.8,0.05],[0.25,0.25,0.5]],dtype=float)
for i in range(10):
    matrix *= matrix
    print("Current round:", i+1)
    print(matrix)
```
